### Importing Dataset

```python
import csv
with open('insurance.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    patient_data = {}
    id_patient = 1
    for patient in reader:
        patient_data[id_patient] = patient
        id_patient += 1
```

### Type Casting Data

```python
for patient in patient_data.values():
    patient['age'] = int(patient['age'])
    patient['bmi'] = float(patient['bmi'])
    patient['children'] = int(patient['children'])
    patient['charges'] = float(patient['charges'])
```

### Testing

```python
print('Patients Data')
print('-'*30)
for id_p, patient in patient_data.items():
    print(f'ID: {id_p}')
    print(f"Age: {patient['age']}")
    print(f"Sex: {patient['sex']}")
    print(f"BMI: {patient['bmi']}")
    print(f"Children: {patient['children']}")
    print(f"Smoker?: {patient['smoker']}")
    print(f"Region: {patient['region']}")
    print(f"Charges: $ {patient['charges']}")
    print('-'*30)
```
