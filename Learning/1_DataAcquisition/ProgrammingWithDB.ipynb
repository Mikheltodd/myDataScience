```python
import sqlite3
import pandas as pd
from numpy.random import uniform

random_numbers = uniform(low=10.0, high=25.0, size=1000)
connection_original = sqlite3.connect("original_pressures.db")
cursor_original = connection_original.cursor()
cursor_original.execute("CREATE TABLE Pressure (reading_p float not null)")

query = "INSERT INTO Pressure (reading_p) VALUES (?)"
for number in random_numbers:
    cursor_original.execute(query, [number])

cursor_original.execute("SELECT * FROM Pressure;")
results = cursor_original.fetchall()

cursor_original.close()
connection_original.commit()
connection_original.close()
```

```python
connection_backup = sqlite3.connect("pressures_20.db")
cursor_backup = connection_backup.cursor()
cursor_backup.execute("CREATE TABLE Pressure (reading_p float not null)")
query = "INSERT INTO Pressure (reading_p) VALUES (?);"

for entry in results:
    # number is saved in first column of the table
    if entry[0] > 20.0:
        cursor_backup.execute(query, entry)

cursor_backup.close()
connection_backup.commit()
connection_backup.close()
```

```python
connection = sqlite3.connect("original_pressures.db")
cursor = connection.cursor()
df = pd.read_sql_query('''SELECT * from Pressure''', connection)

cursor.close()
connection.commit()
connection.close()

connection_2 = sqlite3.connect("pressures_20.db")
cursor_2 = connection_2.cursor()
df_2 = pd.read_sql_query('''SELECT * from Pressure''', connection_2)

cursor_2.close()
connection_2.commit()
connection_2.close()
```
