## Mean, Median and Mode

```python
# Import packages
import numpy as np
import pandas as pd

# Import matplotlib pyplot
from matplotlib import pyplot as plt

# Read in transactions data
greatest_books = pd.read_csv("top-hundred-books.csv")

# Save transaction times to a separate numpy array
author_ages = greatest_books['Ages']
```

### Mean

```python
# Use numpy to calculate the average age of the top 100 authors
average_age = np.average(author_ages)

print("The average age of the 100 greatest authors, according to Le Monde is: " + str(average_age))

# Plot the figure
plt.hist(author_ages, range=(10, 80), bins=14,  edgecolor='black')
plt.title("Age of Top 100 Novel Authors at Publication")
plt.xlabel("Publication Age")
plt.ylabel("Count")
plt.axvline(average_age, color='r', linestyle='solid', linewidth=2, label="Mean")
plt.legend()

plt.show()
```

### Median

```python
# Use numpy to calculate the median age of the top 100 authors
median_age = np.median(author_ages)

# Plot the figure
plt.hist(author_ages, range=(10, 80), bins=14,  edgecolor='black')
plt.title("Age of Top 100 Novel Authors at Publication")
plt.xlabel("Publication Age")
plt.ylabel("Count")
plt.axvline(average_age, color='r', linestyle='solid', linewidth=2, label="Mean")
plt.axvline(median_age, color='y', linestyle='solid', linewidth=2, label="Median")
plt.legend()

plt.show()
```

### Mode

```python
from scipy import stats

mode_age = max(stats.mode(author_ages))

# Plot the figure
plt.hist(author_ages, range=(10, 80), bins=14,  edgecolor='black')
plt.title("Author Ages at Publication")
plt.xlabel("Publication Age")
plt.ylabel("Count")
plt.axvline(average_age, color='r', linestyle='solid', linewidth=3, label="Mean")
plt.axvline(median_age, color='y', linestyle='dotted', linewidth=3, label="Median")
plt.axvline(mode_age, color='orange', linestyle='dashed', linewidth=3, label="Mode")
plt.legend()

plt.show()
```

### Central Tendency for Housing Data

```python
# Import packages
# import numpy as np
# import pandas as pd
# from scipy import stats

# Read in housing data
# In this project, we only care about the price of apartments, so we saved the price of apartments in each borough
brooklyn_one_bed = pd.read_csv('brooklyn-one-bed.csv')
brooklyn_price = brooklyn_one_bed['rent']

manhattan_one_bed = pd.read_csv('manhattan-one-bed.csv')
manhattan_price = manhattan_one_bed['rent']

queens_one_bed = pd.read_csv('queens-one-bed.csv')
queens_price = queens_one_bed['rent']

print(brooklyn_price.head())
print(manhattan_price.head())
print(queens_price.head())
```

```python
# Mean calculations below 
brooklyn_mean = np.average(brooklyn_price)
manhattan_mean = np.average(manhattan_price)
queens_mean = np.average(queens_price)
```

```python
# Median calculations below
brooklyn_median = np.median(brooklyn_price)
manhattan_median = np.median(manhattan_price)
queens_median = np.median(queens_price)
```

```python
# Mode calculations below
brooklyn_mode = stats.mode(brooklyn_price)
manhattan_mode = stats.mode(manhattan_price)
queens_mode = stats.mode(queens_price)
```

```python
# Plot the figure
plt.hist(brooklyn_price, bins=10,  edgecolor='black')
plt.title("Brooklyn Prices")
plt.xlabel("Prices")
plt.ylabel("Count")
plt.axvline(brooklyn_mean, color='r', linestyle='solid', linewidth=3, label="Mean")
plt.axvline(brooklyn_median, color='y', linestyle='dotted', linewidth=3, label="Median")
plt.axvline(max(brooklyn_mode), color='orange', linestyle='dashed', linewidth=3, label="Mode")
plt.legend()

plt.show()
```
